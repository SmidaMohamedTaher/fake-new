```python
import nltk
from nltk.corpus import stopwords

# تحميل قائمة كلمات التوقف العربية
try:
    stopwords.words('arabic')
except LookupError:
    nltk.download('stopwords')

arabic_stopwords = set(stopwords.words('arabic'))

# تعريف الدالة التي تنظف النص (كما في الشرح السابق)
def clean_arabic_text(text):
    if not isinstance(text, str): # تأكد من أن القيمة نص قبل المعالجة
        return text

    words = text.split()
    filtered_words = []

    for word in words:
        # أ. تنظيف الكلمة من علامات الترقيم
        word = re.sub(r'[^\w\s]', '', word)

        # ب. الإزالة الأساسية لكلمات التوقف
        if word not in arabic_stopwords and len(word) > 1:
            # ج. معالجة الـ التعريف وحروف الجر الملتصقة في بداية الكلمة
            stemmed_word = word
            if word.startswith('ال'):
                stemmed_word = word[2:] # حذف "ال"
            elif word.startswith('و'):
                stemmed_word = word[1:] # حذف "و" الملتصقة

            # د. إعادة فحص الكلمة بعد التعديل
            if stemmed_word not in arabic_stopwords and stemmed_word != 'ب':
                filtered_words.append(stemmed_word)

    return ' '.join(filtered_words)
```
